## 파싱
* 어떤 페이지(문서, html 등)에서 내가 원하는 데이터를 특정 패턴이나 순서로 추출해 가공하는 것을 의미
* (참고로, 크롤링(crawling)은 여러 웹 사이트를 돌아다니며 홈페이지의 정보들을 수집하고 검색 대상의 색인으로 포함시키는 기술로 정의됨

```python
from bs4 import BeautifulSoup as bs
from pprint import pprint
import requests

html = requests.get('https://search.naver.com/search.naver?query=날씨')
#pprint(html.text)

soup = bs(html.text,'html.parser')
```

### 요소 1개 찾기(find)

```python
data1 = soup.find('div',{'class':'detail_box'})
pprint(data1)
```

### 요소 모두 찾기(findAll)
find와 사용방법이 똑같으나 find는 처음 매칭된 1개만, findAll은 매칭된 모든 것을 반환하며 리스트로 반환

```python
data2 = data1.findAll('dd')
pprint(data2)
```

### 내부 텍스트 추출
가운데에 있는 실질적인 데이터(숫자와 단위) 부분만 추출

```python
fine_dust = data2[0].find('span',{'class':'num'})
print(fine_dust)
```

 * 내부 텍스트만 골라내도록 .text를 이용한다
 * <태그>텍스트에서 텍스트만 골라내는 방법

```python
fine_dust = data2[0].find('span',{'class':'num'}).text
print(fine_dust)
```

### 초미세먼지 추출
* 여태까지 다룬 것을 응용하여 초미세먼지를 출력 
* data2 변수에서 미세먼지는 0번 인덱스, 초미세먼지는 1번 인덱스

```python
ultra_fine_dust = data2[1].find('span',{'class':'num'}).text
print(ultra_fine_dust)
```
